# Milestone: Creating NumPy Arrays from Python Lists

This milestone focuses on creating NumPy arrays from Python lists, the first step toward efficient numerical computing in Data Science.

## 1. Understanding NumPy Arrays vs Python Lists

First, we import NumPy and compare traditional Python list behavior to NumPy array behavior.

```python
import numpy as np

python_list = [1, 2, 3, 4]
print("Python List:", python_list)
print("Python List * 2:", python_list * 2)
print("Notice how the list duplicates elements instead of multiplying them mathematically.")
```

## 2. Creating NumPy Arrays from Lists

Here we convert Python lists into 1D and multi-dimensional NumPy arrays.

```python
# 1-Dimensional Array
list_1d = [10, 20, 30, 40]
array_1d = np.array(list_1d)

print("1D NumPy Array:")
print(array_1d)

print("\n")

# 2-Dimensional Array (from nested list)
list_2d = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
array_2d = np.array(list_2d)

print("2D NumPy Array:")
print(array_2d)
```

## 3. Inspecting Array Properties

Let's look at the structure and data types of the `array_2d`.

```python
print("Properties of array_2d:")
print(f"- Shape (rows, columns): {array_2d.shape}")
print(f"- Data Type (dtype): {array_2d.dtype}")
print(f"- Dimensionality (ndim): {array_2d.ndim}")
print(f"- Total Elements (size): {array_2d.size}")
```

## 4. Basic Operations on Arrays

A demonstration of element-wise mathematical operations on numeric arrays.

```python
base_array = np.array([5, 10, 15, 20])
print(f"Base Array: {base_array}")

# Element-wise addition
added_array = base_array + 100
print(f"Element-wise Addition (+100): {added_array}")

# Element-wise multiplication
multiplied_array = base_array * 2
print(f"Element-wise Multiplication (*2): {multiplied_array}")

# Operations between arrays
array_b = np.array([1, 2, 3, 4])
sum_arrays = base_array + array_b
print(f"Array + Array (Element-wise Addition): {sum_arrays}")
```
