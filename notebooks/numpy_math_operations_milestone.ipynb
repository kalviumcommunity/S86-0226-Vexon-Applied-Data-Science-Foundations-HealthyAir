# Milestone: Performing Basic Mathematical Operations on NumPy Arrays

This notebook demonstrates how NumPy allows mathematical operations to be applied across entire arrays instantly, bypassing the need for Python loops.

## 1. Element-Wise Array Operations

Let's see how standard arithmetic operators work continuously across corresponding elements of matching arrays.

```python
import numpy as np

array_a = np.array([10, 20, 30, 40])
array_b = np.array([2, 4, 6, 8])

print("Array A:", array_a)
print("Array B:", array_b)
print("--------------------------")
print("Addition (A + B):      ", array_a + array_b)
print("Subtraction (A - B):   ", array_a - array_b)
print("Multiplication (A * B):", array_a * array_b)
print("Division (A / B):      ", array_a / array_b)
```

## 2. Scalar Operations on Arrays

A scalar is a single numerical value. When applied to an array, the operation performs element-wise computation across everything.

```python
print("Original Array:", array_a)
print("\n")
print("Array + 100:  ", array_a + 100)
print("Array * 5:    ", array_a * 5)
print("Array ** 2 (Squared):", array_a ** 2)
```

## 3. Comparing NumPy Arrays and Python Lists

Notice why built-in Python list behavior causes headaches in math-heavy programming—Python just concatenates lists, while NumPy actually performs math.

```python
python_list_1 = [1, 2, 3]
python_list_2 = [4, 5, 6]

np_array_1 = np.array([1, 2, 3])
np_array_2 = np.array([4, 5, 6])

print("PYTHON LIST ADDITION:\n", python_list_1 + python_list_2)
print("\nNUMPY ARRAY ADDITION:\n", np_array_1 + np_array_2)
```

## 4. Avoiding Common Mistakes

Mistakes usually happen when array shapes are completely incompatible. You can't mathematically add an array of shape `(3,)` directly to an array of shape `(4,)`.

```python
array_x = np.array([1, 2, 3])
array_y = np.array([10, 20, 30, 40])

print("Shape X:", array_x.shape)
print("Shape Y:", array_y.shape)
print("\nIf we execute `array_x + array_y`, NumPy will throw a ValueError")
print("because it cannot broadcast shapes (3,) and (4,) together.")
```
