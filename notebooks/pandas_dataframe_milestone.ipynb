# Creating Pandas DataFrames from Dictionaries and Files

This milestone focuses on creating Pandas DataFrames from dictionaries and files, which is a core skill for working with tabular data.

## 1. Understanding Pandas DataFrames

A DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects.

```python
import pandas as pd
```

## 2. Creating DataFrames from Dictionaries

We can create a structured DataFrame programmatically by passing a dictionary of lists to pandas.

```python
data_dict = {
    'Product': ['Laptop', 'Tablet', 'Smartphone', 'Monitor'],
    'Price': [1200, 400, 800, 300],
    'Stock': [50, 100, 150, 75]
}

df_from_dict = pd.DataFrame(data_dict)
print("DataFrame from Dictionary:")
display(df_from_dict)
```

## 3. Creating DataFrames from Files

Most real-world data comes from files like CSVs. Here we will load a sample CSV file.

```python
# Loading from the raw data folder in our workspace structure
file_path = '../data/raw/sample_data.csv'
df_from_file = pd.read_csv(file_path)

print("DataFrame from CSV File:")
display(df_from_file)
```

## 4. Inspecting DataFrame Structure

Once data is loaded, it is important to inspect its shape, columns, and variable types to prevent silent errors.

```python
print("--- First Few Rows ---")
display(df_from_file.head(3))

print("\n--- Column Names ---")
print(df_from_file.columns.tolist())

print("\n--- Shape (rows, columns) ---")
print(df_from_file.shape)

print("\n--- Data Types ---")
print(df_from_file.dtypes)
```
