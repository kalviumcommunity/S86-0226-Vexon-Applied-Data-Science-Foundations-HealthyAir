# Loading CSV Data into Pandas DataFrames

This milestone focuses on loading CSV (Comma-Separated Values) data into Pandas DataFrames. Correct data loading prevents silent errors and ensures your analysis starts on a solid foundation.

## 1. Understanding CSV Files

A CSV file is a plain text file that stores tabular data (numbers and text). Each line of the file is a data record (row). Each record consists of one or more fields, separated by commas (columns). The first row often contains the headers (column names).

```python
import pandas as pd
```

## 2. Loading CSV Files into Pandas

We use the `pd.read_csv()` function to bring CSV data into Python. Pandas automatically assumes the first row contains our column headers.

```python
# Loading a clean standard CSV file
file_path = '../data/raw/clean_data.csv'
df = pd.read_csv(file_path)

print("Data loaded successfully!")
display(df)
```

## 3. Inspecting Loaded Data

Always inspect your data immediately after loading to verify the structure, column names, and row counts.

```python
print("--- First 3 Rows ---")
display(df.head(3))

print("\n--- Column Names ---")
print(df.columns.tolist())

print("\n--- Row and Column Counts ---")
print(f"Rows: {len(df)}, Columns: {len(df.columns)}")

print("\n--- Overall DataFrame Information ---")
df.info()
```

## 4. Recognizing Common Loading Issues

Sometimes CSV files are imperfect. They might have descriptive text at the top, or extra commas that throw off the column alignment. Let's load a 'messy' CSV to see what errors look like, demonstrating why inspection is critical.

```python
# Loading a messy file with metadata at the top and uneven columns
messy_path = '../data/raw/messy_data.csv'

try:
    df_messy = pd.read_csv(messy_path)
    print("Wait, what happened to our columns? Look at the headers:")
    display(df_messy.head())
except Exception as e:
    print("An error occurred during loading!\n", e)
    
# Notice how the headers became "Report Generated Automatically"
# This shows why we MUST inspect data after loading to catch these format issues early.
```
