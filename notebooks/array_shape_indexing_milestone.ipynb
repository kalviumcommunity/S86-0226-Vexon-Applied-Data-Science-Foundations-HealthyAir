# Milestone: Understanding Array Shape, Dimensions, and Index Positions

This milestone focuses on understanding array shape, dimensions, and index positions in NumPy. Correctly navigating arrays is important to safely construct code, prevent index errors, and avoid dimension misinterpretations.

## 1. Understanding Array Shape & 2. Understanding Dimensions

Here, we inspect shapes (rows and columns) and dimensions (1D vs 2D).

```python
import numpy as np

# 1-Dimensional Array
array_1d = np.array([10, 20, 30, 40, 50])
print("1D Array:", array_1d)
print("Shape:", array_1d.shape, "(5 elements,)")
print("Dimensions (ndim):", array_1d.ndim)

print("\n")

# 2-Dimensional Array
array_2d = np.array([
    [100, 200, 300],
    [400, 500, 600],
    [700, 800, 900]
])
print("2D Array (Grid/Table format):")
print(array_2d)
print("Shape:", array_2d.shape, "(3 rows, 3 columns)")
print("Dimensions (ndim):", array_2d.ndim)
```

## 3. Accessing Elements Using Index Positions

Indices in Python start at `0`. In 2D arrays, the format is `[row_index, column_index]`.

```python
# Accessing 1D Arrays
print("First Element (Index 0):", array_1d[0])
print("Third Element (Index 2):", array_1d[2])
print("Last Element  (Index -1):", array_1d[-1])

print("\n")

# Accessing 2D Arrays
print("Element at Row 0, Col 0 (Top-left):", array_2d[0, 0])
print("Element at Row 1, Col 2:", array_2d[1, 2])
print("Element at Row 2, Col 1 (Bottom row, middle column):", array_2d[2, 1])
```

## 4. Visualizing Array Layout

A mapping of how data actually relates to standard zero-based index logic.

```python
print("Row 0:  [100, 200, 300]   <-- indices (0,0), (0,1), (0,2)")
print("Row 1:  [400, 500, 600]   <-- indices (1,0), (1,1), (1,2)")
print("Row 2:  [700, 800, 900]   <-- indices (2,0), (2,1), (2,2)")
```
