# Project Folder Structure Example
This notebook shows a recommended, simple project folder layout and explains where to place files. Use this as the doc you open when starting the project.

## Standard Folders
- `data/raw`: original, immutable data files.
- `data/processed`: cleaned or derived datasets for analysis.
- `notebooks`: exploratory and reporting notebooks.
- `src`: reusable code modules (functions, classes).
- `outputs/figures`: generated figures and plots.
- `outputs/tables`: exported tables or CSVs.
- `models`: saved model artifacts.
- `docs`: documentation and notes.
- `scripts`: small runnable scripts.
- `tests`: unit or integration tests.
- `references`: papers, links, and external resources.
