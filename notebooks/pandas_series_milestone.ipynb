# Creating Pandas Series from Lists and Arrays

This milestone focuses on creating Pandas Series from Python lists and NumPy arrays.

## 1. Understanding Pandas Series

A Pandas Series is a 1D labeled data structure. Unlike raw NumPy arrays, a Series has an explicit index (labels) attached to the values.

```python
import pandas as pd
import numpy as np
```

## 2. Creating a Series from Python Lists

We can convert a Python list into a Pandas Series. Notice the automatic index creation (0, 1, 2...).

```python
my_list = [10, 20, 30, 40, 50]
series_from_list = pd.Series(my_list)

print("Series from List:")
print(series_from_list)
```

## 3. Creating a Series from NumPy Arrays

We can also convert a NumPy array into a Pandas Series. Data types are preserved.

```python
my_array = np.array([1.5, 2.5, 3.5, 4.5])
series_from_array = pd.Series(my_array)

print("Series from NumPy Array:")
print(series_from_array)
```

## 4. Understanding Index and Values

A Series consists of fundamental components: `values` and `index`. We can also create custom labels and access data using them.

```python
# Inspecting index and values
print("Values:", series_from_list.values)
print("Index:", series_from_list.index)

print("\n--- Custom Labels ---")
# Giving meaningful labels to the data
custom_labels = ['A', 'B', 'C', 'D']
labeled_series = pd.Series(my_array, index=custom_labels)
print("Labeled Series:\n", labeled_series)

print("\n--- Accessing Data ---")
print("Access by label ('B'):", labeled_series['B'])
print("Access by positional index (1):", labeled_series.iloc[1])
```
