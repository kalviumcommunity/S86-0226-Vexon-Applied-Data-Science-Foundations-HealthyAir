# Markdown Milestone: Documenting Notebooks
This notebook demonstrates how to write clear Markdown in Jupyter Notebooks: headings, lists, inline code, and code blocks.

## 1. Writing Headings
Use headings to create a clear structure. Examples:
# Top-level heading
## Second-level heading
### Third-level heading
Keep headings short and descriptive.

## 2. Creating Lists
- Use unordered lists for general points.
- Keep items short and focused.

1. Use ordered lists for steps.
2. Numbered steps show sequence clearly.

Use lists to document assumptions, steps, or results.

## 3. Inline Code and Code Blocks
Use inline code for variable names or functions, e.g., `df`, `pd.read_csv()`.
Use fenced code blocks for longer snippets:
```python
# Example: simple function
def square(x):
    return x * x
```
Do not duplicate runnable code unnecessarily—explain intent in Markdown and keep runnable code in code cells.

````python
# Example code cell: simple calculation
x = 5
print('Square of', x, 'is', x**2)
````

## 4. Combining Markdown and Code Cells Effectively
- Put a short Markdown cell before a code cell to explain *why* you're running the code.
- Put a short Markdown cell after code to interpret results.

Example: the next code cell creates a tiny DataFrame and shows it. The Markdown above explains intent; the code cell runs the example.

````python
# Create a small DataFrame to illustrate combining text and code
import pandas as pd
df = pd.DataFrame({'name': ['Alice', 'Bob'], 'score': [85, 92]})
df
````
