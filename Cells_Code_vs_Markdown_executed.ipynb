# Code vs Markdown Cells

This notebook demonstrates the difference between Code cells and Markdown cells and shows when to use each.

## Purpose

Show simple examples of Code cells (which execute) and Markdown cells (which explain).

## Instructions

- Run the Code cells to see outputs.
- Edit or convert cells as needed (press `M` to convert to Markdown, `Y` to convert to Code in command mode).

```python
# Code cell example 1: simple calculation
x = 5
y = 7
print('Sum:', x + y)
```

The cell above is a Code cell — it runs Python and prints the result. Use Code cells for computations, data loading, and plotting.

```python
# Code cell example 2: loop output
for i in range(3):
    print('i =', i)
```

The cell above shows a multi-line Code cell. Use short, focused Code cells rather than dumping all logic into one long cell.

## Switching cell types

- In command mode: press `M` to convert to Markdown, `Y` to convert to Code.
- From the menu: Cell → Cell Type → Markdown/Code.
Use Markdown for titles, explanations, and notes; use Code for runnable logic.
