## 1. Import Required Libraries

We will use Pandas for data handling. If you want to create more complex sample data, you can also import NumPy.

```python
import pandas as pd

data = {
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bangalore'],
    'PM2.5': [110, 90, 70, 80, 60],
    'PM10': [200, 180, 150, 160, 140],
    'NO2': [40, 35, 30, 32, 28]
}
df = pd.DataFrame(data)
df.head()
```

## 2. Load a Sample DataFrame

Let's create a small DataFrame to inspect. In real projects, you would typically load data from a file using `pd.read_csv()`, but here we'll use synthetic data for clarity.

```python
# 1. Selecting Columns by Name
# Select a single column
city_col = df['City']
print('Single column (City):')
print(city_col)

# Select multiple columns
cols = df[['City', 'PM2.5']]
print('\nMultiple columns (City, PM2.5):')
print(cols)

# Check type
print('\nType of single column selection:', type(city_col))
print('Type of multiple column selection:', type(cols))
```

## 3. Preview Data with `head()`

The `head()` method shows the first few rows of the DataFrame (default is 5). This helps you quickly check the data's structure, column names, and sample values.

```python
# 2. Selecting Rows by Position
# Select the first row (position 0)
first_row = df.iloc[0]
print('First row by position:')
print(first_row)

# Select first three rows
first_three = df.iloc[:3]
print('\nFirst three rows:')
print(first_three)
```

You can also specify the number of rows to preview. For example, `df.head(3)` shows the first 3 rows.

```python
# 3. Selecting Rows by Label
# Set City as index for label-based selection
df_label = df.set_index('City')

# Select row by label
row_mumbai = df_label.loc['Mumbai']
print('Row for Mumbai by label:')
print(row_mumbai)

# Slice rows by label
slice_rows = df_label.loc['Delhi':'Kolkata']
print('\nRows from Delhi to Kolkata:')
print(slice_rows)
```

## 4. Inspect Structure with `info()`

The `info()` method provides details about the DataFrame's structure, including column names, data types, non-null counts, and memory usage. This helps you spot missing values and understand the data types for each column.

```python
# 4. Selecting Rows and Columns Together
# Select specific rows and columns by position
subset_pos = df.iloc[1:4, [0, 2]]  # Rows 1-3, columns City and PM10
print('Rows 1-3, columns City and PM10:')
print(subset_pos)

# Select specific rows and columns by label
subset_label = df_label.loc[['Delhi', 'Bangalore'], ['PM2.5', 'NO2']]
print('\nDelhi and Bangalore, columns PM2.5 and NO2:')
print(subset_label)
```

Look for columns with missing values or unexpected data types. This helps you decide if cleaning is needed before analysis.

## 5. Summarize Numeric Data with `describe()`

The `describe()` method provides summary statistics for all numeric columns, including count, mean, standard deviation, min, max, and percentiles. This helps you understand the distribution and spot potential outliers.

```python
# 5. Summary: When to Use Each Approach
print('---Summary---')
print('Use column names for clear, explicit column selection.')
print('Use iloc for position-based row/column selection (quick slicing, zero-based).')
print('Use loc for label-based selection (index values, inclusive slicing).')
print('Combine row and column selection for precise data extraction.')
print('Always verify your selection to avoid mistakes.')
```

Look for unusual values (potential outliers) and check if the statistics make sense. Remember, `describe()` only summarizes numeric columns by default.

## 6. When to Use Each Inspection Method

- Use `head()` for a quick visual preview of your data and to check alignment.
- Use `info()` to understand the structure, data types, and spot missing values.
- Use `describe()` to get a statistical summary of numeric columns and check for outliers.

**Best practice:** Always use all three methods before starting any data cleaning or analysis. This ensures you understand your dataset and can catch issues early.

## Detecting Missing Values in DataFrames
This section demonstrates how to detect, summarize, and inspect missing values in a Pandas DataFrame, as required by the milestone assignment.

```python
# 1. Understanding Missing Values
import numpy as np
import pandas as pd

# Example DataFrame with missing values
data_missing = {
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bangalore'],
    'PM2.5': [110, np.nan, 70, 80, None],
    'PM10': [200, 180, np.nan, 160, 140],
    'NO2': [40, 35, 30, None, 28]
}
df_missing = pd.DataFrame(data_missing)
df_missing
```

```python
# 2. Detecting Missing Values
# Detect missing values (returns boolean DataFrame)
missing_mask = df_missing.isnull()
print('Missing value mask:')
print(missing_mask)

# Identify which columns contain missing data
cols_with_missing = df_missing.columns[df_missing.isnull().any()].tolist()
print('\nColumns with missing data:', cols_with_missing)

# Inspect missing values row-wise
rows_with_missing = df_missing.isnull().any(axis=1)
print('\nRows with missing data:')
print(df_missing[rows_with_missing])
```

```python
# 3. Counting Missing Values
# Count missing values per column
missing_per_col = df_missing.isnull().sum()
print('Missing values per column:')
print(missing_per_col)

# Proportion of missing values per column
missing_prop = df_missing.isnull().mean()
print('\nProportion of missing values per column:')
print(missing_prop)
```

```python
# 4. Inspecting Rows with Missing Data
# Identify rows containing missing values
rows_with_missing = df_missing[df_missing.isnull().any(axis=1)]
print('Rows with missing data:')
print(rows_with_missing)

# Inspect patterns of missingness
print('\nPattern of missingness:')
print(rows_with_missing.isnull())
```

### Summary: Why Detecting Missing Values Matters
- Always check for missing data before analysis or cleaning.
- Missing values can affect statistics, visualizations, and model results.
- Early detection ensures trustworthy, intentional data cleaning and analysis.
- Use Pandas methods like `isnull()`, `sum()`, and boolean indexing for robust detection.
