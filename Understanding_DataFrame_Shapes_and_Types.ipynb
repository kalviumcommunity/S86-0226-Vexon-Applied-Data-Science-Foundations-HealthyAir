## 1. Import Required Libraries and Load Data

Let's import pandas and create a sample DataFrame. In real projects, you might load data from a CSV file.

```python
import pandas as pd

# Sample data with mixed types and a string-number column
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 29],
    'Score': ['88.5', '92.0', '79.5', '85.0', None],  # Note: numbers as strings and a missing value
    'Passed': [True, True, False, True, True],
    'JoinDate': ['2022-01-10', '2022-02-15', '2022-03-20', '2022-04-25', '2022-05-30']
}
df = pd.DataFrame(data)
df
```

## 2. Inspect DataFrame Shape

The `.shape` attribute shows the number of rows and columns in the DataFrame. The first value is the number of rows (records), and the second is the number of columns (features/variables).

```python
# Show the shape of the DataFrame
df.shape
```

The output is a tuple: (number of rows, number of columns). For example, (5, 5) means 5 records and 5 features.

## 3. Interpret Rows and Columns

Rows represent individual observations (records), and columns represent features (variables/attributes). Let's count them explicitly.

```python
# Number of rows (observations)
num_rows = len(df)
print('Number of rows:', num_rows)

# Number of columns (features)
num_cols = len(df.columns)
print('Number of columns:', num_cols)

# List column names
df.columns.tolist()
```

## 4. Inspect Column Data Types

The `.dtypes` attribute and `info()` method show the data type of each column. This is essential for understanding what operations are valid and for detecting issues.

```python
# Show data types of each column
df.dtypes
```

```python
# Show more details including non-null counts and memory usage
df.info()
```

## 5. Identify and Interpret Common Data Types

Pandas uses several common data types:
- `int64`: Integer numbers
- `float64`: Decimal numbers
- `object`: Usually text (strings), but can be mixed types
- `bool`: Boolean values (True/False)
- `category`: Categorical data (efficient for repeated values)
- `datetime64`: Dates and times

Let's identify which columns are numeric, categorical, or text in our DataFrame.

```python
# Identify numeric columns
df.select_dtypes(include=['number'])

# Identify object (text) columns
df.select_dtypes(include=['object'])

# Identify boolean columns
df.select_dtypes(include=['bool'])
```

## 6. Detect Type-Related Issues

Sometimes, columns that look numeric are actually stored as strings (object type), or missing values cause mixed types. Let's check for these issues.

```python
# Check if 'Score' column is numeric or object
df['Score'].dtype

# Try converting 'Score' to numeric, see what happens with errors
pd.to_numeric(df['Score'], errors='coerce')

# Check for missing values in the DataFrame
df.isnull().sum()
```

If you see that a numeric column is stored as `object`, or if there are missing values, you may need to clean or convert the data before analysis. Always check these issues early to avoid errors later.
